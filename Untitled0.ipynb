```python
def mostrar_menu():
    print('\n--- } SISTEMA DE GESTIÓN "ADOPTA-UN-AMIGO" ---')
    print("1. Ver estado de alimento")
    print("2. Ver animales listos para adopción")
    print("3. Cambiar nombre del refugio")
    print("4. Calcular costo de vacunación (Opción Matemática)")
    print("5. Salir")

def ejecutar_asistente():
    refugio = "Huellitas Felices"
    continuar = True

    while continuar:
        mostrar_menu()
        opcion = input("Seleccione una opción (1-5): ")

        if opcion == "1":
            print(" Stock de alimento al 85%.")
        elif opcion == "2":
            print(" Hay 12 perros y 8 gatos esperando un hogar.")
        elif opcion == "3":
            refugio = input("Nuevo nombre para el refugio: ")
            print(f" Nombre actualizado: {refugio}")
        elif opcion == "4":
            # --- CÁLCULO MATEMÁTICO ---
            cantidad = int(input("¿Cuántos animales vas a vacunar? "))
            precio_unitario = 1500
            total = cantidad * precio_unitario
            print(f" El costo total de las vacunas es: ${total}")
        elif opcion == "5":
            print(f" Cerrando sistema de {refugio}. ¡Hasta luego!")
            continuar = False
        else:
            print(" Opción no válida.")

ejecutar_asistente()
```
